```python
df = pd.read_csv('sales.csv', encoding = "ISO-8859-1")
```

```python
# Identify missing values using .isnull()
print("Missing values in the dataset:")
print(df.isnull())
```

```python
# Filter rows with missing values
print("\nRows with missing values:")
print(df[df.isnull().any(axis=1)])
```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()

print(df_cleaned)
```

```python
# Standardize to lowercase
df['CUSTOMERNAME'] = df['CUSTOMERNAME'].str.lower()
print(df)
```

```python
# Remove leading and trailing whitespace
df['CUSTOMERNAME'] = df['CUSTOMERNAME'].str.strip()
print(df)
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
print(df.columns)
```

```python
df['ordernumber'] = df['ordernumber'].astype(int)  # Convert age to integer
df['orderdate'] = pd.to_datetime(df['orderdate'])  # Convert to datetime
print(df)
```
